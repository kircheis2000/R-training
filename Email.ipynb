```python
import pandas as pd
import pandasql as pdsql
```

```python
email = pd.read_csv("email.csv")
email
```

```python
pysql = lambda q: pdsql.sqldf(q, globals())
```

```python
sql = """ select
            a1.member_id
            , a1.email as email1
            , a2.email as email2
            
            from email a1
            inner join email a2
            on a1.member_id = a2.member_id and a1.email <> a2.email
            """
df = pysql(sql)
df
```

```python
email_all = email.merge(email, left_on = "member_id", right_on = "member_id", how = "outer").\
            rename(columns = {"email_x": "email1", "email_y": "email2"})
email_all = email_all[email_all.email1 != email_all.email2]
```
